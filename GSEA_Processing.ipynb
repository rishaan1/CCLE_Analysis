```python
import pandas as pd
dep_tbl = pd.read_csv("dep_tbl.csv", sep = '\t')
dep_tbl
```

```python
#This block creates the phenotype labels file in the specific format required for GSEA
phenotype_labels = dep_tbl[["ModelID", "ecDNA Status"]]
# print(len(dep_tbl["ModelID"]))
with open("phenotype_labels.cls", "w") as f:
    f.write(str(len(dep_tbl["ModelID"])) + " " + str(len(dep_tbl["ecDNA Status"].unique())) + " 1\n")
    f.write("# ecDNA- ecDNA+\n")
    for index, row in phenotype_labels.iterrows():
        if row["ecDNA Status"] == "ecDNA+":
            f.write("1 ")
        elif row["ecDNA Status"] == "ecDNA-":
            f.write("0 ")
        
    
    
```

```python
expression_data = dep_tbl.drop('ecDNA Status', axis = 1)
changes = {}
columns = expression_data.columns.tolist()
# print(type(columns))
for col in columns: 
    gene = str(col).split(" ")
    changes[col] = gene[0]
expression_data.rename(columns = changes, inplace = True)
expression_data.head()
expression_data = expression_data.drop("Unnamed:", axis = 1)
expression_data.head()
expression_data = expression_data.transpose()
expression_data.head()
expression_data.insert(loc=0, column='Description', value="na")
expression_data.to_csv("expression_data.csv")
#open this file in Excel, add necessary rows to the top and convert to gct file
```

```python
gene_set = expression_data.columns
gene_set = gene_set.to_list()
gene_set = gene_set[1:]
gene_set =  pd.DataFrame(gene_set, columns=['Genes'])
gene_set.reset_index()
for index, row in gene_set.iterrows():
    string1 = row["Genes"]
    gene = string1.split(" ")
    gene_set.loc[index, "Genes"] = gene
gene_set.to_csv("gene_set.csv")
#recommended to use custom gene sets from GSEA, could create your own 
```
